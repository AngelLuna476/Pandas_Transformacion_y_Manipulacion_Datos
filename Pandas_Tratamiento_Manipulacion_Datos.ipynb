#**PROBLEMA DE NEGOCIO**


---




Se trata de un algoritmo capaz de analizar las características de una propiedad -como comodidades, tamaño, ocupación del espacio en un período determinado- y sugerir al anfitrión un precio a cobrar por tarifas diarias que garantice ganancias en momentos de alta demanda.

#**1. Entendiendo el Problema**


---




```python
#Importación de pandas
import pandas as pd
```

```python
#Lectura y revisión del contenido
datos = pd.read_json('/content/datos_hosting.json')
datos.head()
```

```python
#Acomodo de los datos a la estandarización en pandas del archivo .JSON y revisión
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

#**2. Datos numéricos**


---




```python
#Viazualización de las columnas en el Data Frame que se nombró anteriormente
columnas = list(datos.columns)
columnas
```

```python
#Realización del "explotado" de los datos, motivo para separar los diccionarios anidados en las columnas y su separación
datos = datos.explode(columnas[3:])
datos
```

```python
# Realización de reseteo de los "índices" para que no ocupen más de una sola fila
datos.reset_index(inplace = True, drop = True)
datos.head()
```

```python
# Realización de revisión de los tipos de datos para un mejor tratamiento de los mismos y detectar los errores existentes
datos.info()
```

```python
# Importación de la biblioteca de Numpy para el tratamiento de los datos del tipo númerico
import numpy as np
```

```python
# Conversión del tipo de los datos de "max_hospedes" de OBJECT a INT64 para su mejor tratamiento
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
# Revisión de que el cambio haya quedado guardado
datos.info()
```

```python
# Creación de una variable que contenga las últimas columnas del tipo númerico para su conversión de OBJECT a INT
col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
# Conversión de las columnas almacenadas en la variable a tipo INT
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
# Revisión de que los cambios de hayan guardado para su tratamiento correspondiente
datos.info()
```

```python
# conversión del tipo OBJECT al FLOAT a la columna que presenta datos con punto decimal
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
# Revisión de que los cambios se hayan guardado para su debido tratamiento
datos.info()
```

```python
# Eliminación de caracterés especiales de la columna PRECIO y la eliminación del "ESPACIO" que ocupaba ese caracter especial, así como en numeros a partir de MIL, la eliminación de la separación
#correspondiente
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
# Coversión del tipo de dato OBJECT a FLOAT
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
# Revisión de que los cambios hayan sido guardados
datos.info()
```

```python
# Datos faltantes de camviar los caracteres que son del tipo float del data frame
datos[['cuota_deposito', 'cuota_limpieza']].head(10)
```

```python
# Tratamiento de la <<Eliminación de los caracteres especiales "$" y reemplazarlos con "NADA", así como de las "," por "NADA" y quitar los "Espacios" que se ocupaban,
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].map(lambda x: x.replace('$','').replace(',','').strip())
```

```python
#Se cambia el comando .applymap por el comando .map por las actualizaciones que menciona el colab
```

```python
datos[['cuota_deposito', 'cuota_limpieza']].head(10)
```

```python
# Creación de una varisble para la conversión de las columnasdel tipo OBJECT a FLOAT de los datos que se encuentran en esas columnas
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
# Revisión de que los cambios se hayan guardado par su tratamiento
datos.info()
```

#**3. Datos de texto**


---




```python
# Revisión del Data Frame para inspección de que más cambios se deben de hacer
datos.head()
```

```python
# Cambio de las letras que se encuentran en mayúsculas de la columna a mínusculas para su tratamiento
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
# Revisión de los cambios al vizualizar el Data Frame
datos.head()
```

```python
# Revisión del índice 3169 (se seleccionó de forma aleatoria) para la inspección de caracteres especiales existentes que puedan ocasionar un error al convertirlos en "diccionarios", (Se realizará un tipo de busqueda por palabras para verificar si ciertas palabras afectan el rendimiento de la "Evaluación General")
datos['descripcion_local'][3169]
```

```python
# Realización de una búsqueda más compleja con el comando <<REGEX>> de los caracterés que se quitaran para el diccionario, Valores "Concisos" como letras y números son separados por un "-", caracteres especiales por una "\" para indicar de un cierto punto a otro se eliminarán los datos y reemplazarán (en este caso) por un espacio en blanco
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
# Realización de la búsqueda de los criterios a eliminar siendo estos caracteres especiales debido a la existencia posible de elementos como << one - time >> siendo estos no palabras compuestas en el Inglés
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)',' ',regex=True)
```

```python
# Revisión de los datos y sus cambios en la columna descripcion_local
datos['descripcion_local'].head()
```

```python
# Creación de un tipo "diccionario" por los corchetes que tiene en el incio, el cual ayudará a la realización de búsqueda por letras
datos['descripcion_local'] = datos['descripcion_local'].str.split()
datos.head()
```

```python
# Eliminación de los caracteres especiales de la columna comodidades
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"','', regex=True)
```

```python
# Eliminaciónm de los espacios existentes por los caracteres especiales y cambio de separador de datos a "," y su debida revisión de los cambios realizados que se hayan guardado, asi como de un diccionario
datos['comodidades'] = datos['comodidades'].str.split(',')
datos.head()
```

```python
# Transformamos el texto en letras minúsculas
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.lower()
```

```python
# Substituímos los caracteres especiales
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
# Transformamos el texto en lista, formando el token
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.split()
datos.head()
```
